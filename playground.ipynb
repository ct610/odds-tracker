```python
from data.wordle_words import list_words

def jenis_solver_1(start_word, start_tiles):
    # first get set of possible words
    words = list_words()
    filtered_words = set()
    for word in words:
        # if this is the true word, does this produce the start tiles?
        tiles = ['N', 'N', 'N', 'N', 'N']
        mod_word = list(word)
        for i, char in enumerate(start_word):
            if char == mod_word[i]:
                tiles[i] = 'G'
                mod_word[i] = '.'
        for i, char in enumerate(start_word):
            if char in mod_word:
                if tiles[i] != 'G':
                    j = mod_word.index(char)
                    tiles[i] = 'Y'
                    mod_word[j] = '.'
        if tiles == start_tiles:
            filtered_words.add(word)
    return filtered_words
```

```python
def jenis_solver_2(filtered_words):
    # return word(s) with highest win probability
    best_prob = -1e-4
    best_words = set()
    words = list_words()
    for word in words:
        # if this is guess 2, how does it divide up what we get in guess 1?
        d = {}
        for possible in filtered_words:
            tiles = ['N', 'N', 'N', 'N', 'N']
            mod_word = list(possible)
            for i, char in enumerate(word):
                if char == mod_word[i]:
                    tiles[i] = 'G'
                    mod_word[i] = '.'
            for i, char in enumerate(word):
                if char in mod_word:
                    if tiles[i] != 'G':
                        j = mod_word.index(char)
                        tiles[i] = 'Y'
                        mod_word[j] = '.'
            tiles = "".join(tiles)
            if tiles in d.keys():
                d[tiles] += 1
            else:
                d[tiles] = 1
        prob = len(d) / sum(d.values())
        if prob > best_prob + 1e-4:
            best_words = {word}
            best_prob = prob
        elif prob < best_prob - 1e-4:
            continue
        else:
            best_words.add(word)
    return best_words, best_prob
```

```python
filtered_words = jenis_solver_1(['b', 'r', 'a', 'i', 'n'], ['N', 'N', 'G', 'N', 'N'])
jenis_solver_2(filtered_words)
```

```python
words = list_words()

best_start = ""
best_prob = -0.001
for c, word in enumerate(words):
    if c % 10 == 0:
        print(c)
    possible_start_tiles = {}
    for sol in words:
        if sol == word:
            continue
        mod_word = list(sol)
        tiles = ['N', 'N', 'N', 'N', 'N']
        for i, char in enumerate(word):
            if char == mod_word[i]:
                tiles[i] = 'G'
                mod_word[i] = '.'
        for i, char in enumerate(word):
            if char in mod_word:
                if tiles[i] != 'G':
                    j = mod_word.index(char)
                    tiles[i] = 'Y'
                    mod_word[j] = '.'
        tiles = "".join(tiles)
        if tiles in possible_start_tiles.keys():
            possible_start_tiles[tiles] += 1
        else:
            possible_start_tiles[tiles] = 1
    total_prob = 0
    for tiles in possible_start_tiles.keys():
        filtered_words = jenis_solver_1(list(word), list(tiles))
        _, p = jenis_solver_2(filtered_words)
        total_prob += p * possible_start_tiles[tiles]
    final_prob = total_prob / sum(possible_start_tiles.values())
    if final_prob > best_prob:
        best_start = word
        best_prob = final_prob
print(best_start, best_prob)
```
